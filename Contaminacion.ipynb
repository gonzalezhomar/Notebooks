```python
import folium
from folium.plugins import HeatMap
import pandas as pd
```

```python
HeatMap?
```

```python
df = pd.read_csv('contaminacion.csv', encoding='latin1')
df.dtypes
```

```python
aire=df.loc[df['Aire'] > 0]
aire.shape
```

```python
m=folium.Map(location=[22.307493031549864, -101.86451798709956],
             zoom_start=3,
             min_zoom=5,
             max_zoom=16)
HeatMap(data=aire[['Latitud', 'Longitud', 'Aire']]).add_to(m)
m
```

```python
agua=df.loc[df['Agua'] > 0]
agua.shape
```

```python
m2=folium.Map(location=[22.307493031549864, -101.86451798709956],
             zoom_start=3,
             min_zoom=5,
             max_zoom=16)
HeatMap(data=agua[['Latitud', 'Longitud', 'Agua']]).add_to(m2)
m2
```

```python
suelo=df.loc[df['Suelo'] > 0]
suelo.shape
```

```python
m3=folium.Map(location=[22.307493031549864, -101.86451798709956],
             zoom_start=3,
             min_zoom=5,
             max_zoom=16)
HeatMap(data=suelo[['Latitud', 'Longitud', 'Suelo']]).add_to(m3)
m3
```

```python
otros=df.loc[df['Otros'] > 0]
otros.shape
```

```python
m5=folium.Map(location=[22.307493031549864, -101.86451798709956],
             zoom_start=3,
             min_zoom=5,
             max_zoom=16)
HeatMap(data=otros[['Latitud', 'Longitud', 'Otros']]).add_to(m5)
m5
```

```python
m4=folium.Map(location=[22.307493031549864, -101.86451798709956],
             zoom_start=3,
             min_zoom=5,
             max_zoom=16)
HeatMap(data=aire[['Latitud', 'Longitud', 'Aire']], name='Aire').add_to(m4)
HeatMap(data=agua[['Latitud', 'Longitud', 'Agua']], name='Agua').add_to(m4)
HeatMap(data=suelo[['Latitud', 'Longitud', 'Suelo']], name='Suelo').add_to(m4)
HeatMap(data=otros[['Latitud', 'Longitud', 'Otros']], name='Otros').add_to(m4)
folium.LayerControl().add_to(m4)
m4
```

```python
m4.save('contaminacion.html')
```
