```python
import folium
from folium import plugins
import pandas as pd
import geopandas as gpd
import plotly_express as px
import matplotlib.pyplot as plt
from branca.element import Figure
```

```python
df = pd.read_csv('2020-10-26_datos_urbanos2.csv')
df.dtypes
```

```python
subidas=df.loc[df['SubidasAdulto'] > 0]
subidas=subidas.loc[subidas['Unidad'] == 1]
#subidas = subidas.iloc[0:1000, :]
subidas.shape
```

```python
bajadas=df.loc[df['BajadasAdulto'] > 0]
bajadas=bajadas.loc[bajadas['Unidad'] == 1]
#subidas = subidas.iloc[0:1000, :]
bajadas.shape
```

```python
fig=Figure(width=550,height=350)
m=folium.Map(location=[21.128150, -101.662665])
fig.add_child(m)
```

```python
mapasubidas = folium.map.FeatureGroup()
for lat, lng, in zip(subidas.Latitud, subidas.Longitud):
    mapasubidas.add_child(
        folium.CircleMarker(
        [lat, lng],
        radius=3, # define how big you want the circle markers to be
        color='green'#,
        #fill=True,
        #fill_color='blue',
        #fill_opacity=0.6
        )
    )
```

```python
mapabajadas = folium.map.FeatureGroup()
for lat, lng, in zip(bajadas.Latitud, bajadas.Longitud):
    mapabajadas.add_child(
        folium.CircleMarker(
        [lat, lng],
        radius=2, # define how big you want the circle markers to be
        color='red'#,
        #fill=True,
        #fill_color='black',
        #fill_opacity=0.6
        )
    )
```

```python
fig2=Figure(width=550,height=350)
m2=folium.Map(width=550,height=350,location=[21.12814, -101.6626],zoom_start=12,min_zoom=7,max_zoom=16)
m2.add_child(mapasubidas)
m2.add_child(mapabajadas)
fig2.add_child(m2)
```

```python
fig2.save('index.html')
fig2.save("filename.png")
```
