```python
import pandas as pd
import folium
from folium import plugins
from folium.plugins import MarkerCluster
from branca.element import Figure
```

```python
df_oficinas=pd.read_csv('OficinasRecaudadoras.csv', encoding= 'unicode_escape')
df_oficinas=df_oficinas.dropna(subset=['Longitude'])
df_oficinas.head()
```

```python
oficinas = folium.map.FeatureGroup()
for lat, lng, in zip(df_oficinas.Latitude, df_oficinas.Longitude):
    oficinas.add_child(
        folium.Marker(
        [lat, lng],
        radius=3, # define how big you want the circle markers to be
        color='blue'#,
        #fill=True,
        #fill_color='blue',
        #fill_opacity=0.6
        )
    )
print(oficinas)
```

```python
fig2=Figure(width=550,height=350)
m2=folium.Map(width=550,height=350,location=[20.972658157717838, -101.05460389352301], zoom_start=8)
m2.add_child(oficinas)
#m2.add_child(mapabajadas)
fig2.add_child(m2)
```

```python
fig=Figure(width=550,height=350)
m = folium.Map(location=[20.972658157717838, -101.05460389352301], zoom_start=8)
#folium.LayerControl().add_to(m)
marker_cluster = MarkerCluster().add_to(m)
for row in df_oficinas.itertuples():
    folium.Marker(location=[row.Latitude,row.Longitude],popup="<strong>OFICINA: </strong>"+row.Oficina+"\n <strong>TELEFONO: </strong>"+row.Telefono + "; \n <strong>TITULAR: </strong>"+ row.Titular).add_to(marker_cluster)
fig.add_child(m)
```

```python
fig.save('oficinas_agrupadas.html')
fig2.save('oficinas_puntos.html')
```
